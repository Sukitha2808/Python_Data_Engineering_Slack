# Task
Analyze shipment, delivery, and claim data from "shipments.csv", "delivery_logs.csv", and "claims.csv" to calculate delay duration, flag reorders, and determine claim aging using Python with Pandas.

## Load data

### Subtask:
Load the `shipments.csv`, `delivery_logs.csv`, and `claims.csv` files into pandas DataFrames.


**Reasoning**:
Import pandas and load the three CSV files into separate DataFrames as instructed.



```python
import pandas as pd

df_shipments = pd.read_csv('/content/shipments.csv')
df_delivery = pd.read_csv('/content/delivery_logs.csv')
df_claims = pd.read_csv('/content/claims.csv')
```

## Clean shipment data

### Subtask:
Clean shipment data by converting relevant columns to appropriate data types (e.g., dates to datetime objects) and handling missing values if necessary.


**Reasoning**:
Convert the date columns to datetime objects and check for missing values in the dataframe.



```python
df_shipments['ship_date'] = pd.to_datetime(df_shipments['ship_date'])
df_shipments['delivery_date'] = pd.to_datetime(df_shipments['delivery_date'])
missing_values = df_shipments.isnull().sum()
print("Missing values in df_shipments:")
print(missing_values)
```

## Clean delivery data

### Subtask:
Clean delivery data by converting relevant columns to appropriate data types (e.g., dates to datetime objects) and handling missing values if necessary.


**Reasoning**:
Inspect the dataframe for date columns and check for missing values.



```python
print(df_delivery.info())
print(df_delivery.isnull().sum())
```

## Clean claims data

### Subtask:
Clean claims data by converting relevant columns to appropriate data types (e.g., dates to datetime objects) and handling missing values if necessary.


**Reasoning**:
Convert the specified columns to datetime objects and check for missing values in the claims dataframe.



```python
df_claims['claim_date'] = pd.to_datetime(df_claims['claim_date'])
df_claims['resolved_date'] = pd.to_datetime(df_claims['resolved_date'])
missing_values_claims = df_claims.isnull().sum()
print("Missing values in df_claims:")
print(missing_values_claims)
```

## Merge data

### Subtask:
Merge the three DataFrames based on relevant keys (e.g., shipment ID).


**Reasoning**:
Merge the three dataframes sequentially and display the head of the final merged dataframe.



```python
df_merged = pd.merge(df_shipments, df_delivery, on='shipment_id', how='inner')
df_merged = pd.merge(df_merged, df_claims, on='delivery_id', how='inner')
display(df_merged.head())
```

## Calculate delay duration

### Subtask:
Calculate the difference between the delivery date and the estimated delivery date.


**Reasoning**:
Calculate the actual delivery duration and the delay duration, then display the head of the merged DataFrame.



```python
df_merged['calculated_delivery_duration'] = (df_merged['delivery_date'] - df_merged['ship_date']).dt.days
df_merged['delay_duration'] = df_merged['calculated_delivery_duration'] - df_merged['delivery_duration_days']
display(df_merged.head())
```

## Flag reorders

### Subtask:
Flag reorders by identifying shipments that are reorders based on the data.


**Reasoning**:
Identify reorders by grouping the data and flagging groups with more than one shipment as reorders.



```python
df_merged['is_reorder'] = df_merged.groupby(['origin_warehouse', 'destination_city', 'product_id'])['shipment_id'].transform('size') > 1
display(df_merged.head())

```

## Calculate claim aging

### Subtask:
Calculate the difference between the claim date and the resolution date (if available) or the current date.


**Reasoning**:
Calculate the claim aging by finding the difference between the resolved date and claim date, or the current date and claim date if the resolved date is missing. Then display the head of the dataframe.



```python
from datetime import datetime

current_date = datetime.now()
df_merged['claim_aging'] = (df_merged['resolved_date'].fillna(current_date) - df_merged['claim_date']).dt.days
display(df_merged.head())

```

## Summary:

### Data Analysis Key Findings

*   There were no missing values in the `shipments`, `delivery_logs`, or `claims` datasets after loading.
*   Date columns (`ship\_date`, `delivery\_date`, `claim\_date`, `resolved\_date`) were successfully converted to datetime objects.
*   The datasets were successfully merged into a single DataFrame (`df_merged`) using 'shipment\_id' and 'delivery\_id'.
*   A `delay\_duration` column was calculated, representing the difference in days between the actual and estimated delivery dates.
*   A boolean `is\_reorder` column was added to flag shipments considered reorders based on repeated shipments from the same origin warehouse to the same destination city for the same product.
*   A `claim\_aging` column was calculated, showing the number of days between the claim date and the resolution date (or the current date if not resolved).

## Define sqlalchemy models

### Subtask:
Create SQLAlchemy models for shipments, delivery logs, and claims to represent the database schema.


**Reasoning**:
Define SQLAlchemy models for Shipment, DeliveryLog, and Claim including their respective columns, data types, primary keys, and foreign keys.



```python
from sqlalchemy import create_engine, Column, Integer, String, Date, ForeignKey, Boolean, Float
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import relationship

Base = declarative_base()

class Shipment(Base):
    __tablename__ = 'shipments'

    shipment_id = Column(String, primary_key=True)
    origin_warehouse = Column(String)
    destination_city = Column(String)
    ship_date = Column(Date)
    delivery_date = Column(Date)
    product_id = Column(String)
    quantity = Column(Integer)
    freight_cost = Column(Float)

class DeliveryLog(Base):
    __tablename__ = 'delivery_logs'

    delivery_id = Column(String, primary_key=True)
    shipment_id = Column(String, ForeignKey('shipments.shipment_id'))
    carrier = Column(String)
    status = Column(String)
    delivery_duration_days = Column(Integer)
    damage_flag = Column(Boolean)
    proof_of_delivery_status = Column(String)

    shipment = relationship("Shipment")

class Claim(Base):
    __tablename__ = 'claims'

    claim_id = Column(String, primary_key=True)
    delivery_id = Column(String, ForeignKey('delivery_logs.delivery_id'))
    reason = Column(String)
    amount_claimed = Column(Float)
    claim_status = Column(String)
    claim_date = Column(Date)
    resolved_date = Column(Date)

    delivery_log = relationship("DeliveryLog")
```

## Define pydantic models

### Subtask:
Create Pydantic models for data validation and serialization for API requests and responses.


**Reasoning**:
Define Pydantic models for data validation and serialization for API requests and responses, including configuration for ORM mode.



```python
from pydantic import BaseModel
from datetime import date
from typing import Optional

class ShipmentBase(BaseModel):
    shipment_id: str
    origin_warehouse: str
    destination_city: str
    ship_date: date
    delivery_date: date
    product_id: str
    quantity: int
    freight_cost: float

    class Config:
        orm_mode = True

class DeliveryLogBase(BaseModel):
    delivery_id: str
    shipment_id: str
    carrier: str
    status: str
    delivery_duration_days: int
    damage_flag: bool
    proof_of_delivery_status: str

    class Config:
        orm_mode = True

class ClaimBase(BaseModel):
    claim_id: str
    delivery_id: str
    reason: str
    amount_claimed: float
    claim_status: str
    claim_date: date
    resolved_date: Optional[date] = None

    class Config:
        orm_mode = True

class ShipmentCreate(ShipmentBase):
    pass

class DeliveryLogCreate(DeliveryLogBase):
    pass

class ClaimCreate(ClaimBase):
    pass

class ClaimsSummaryResponse(BaseModel):
    claim_id: str
    delivery_id: str
    reason: str
    amount_claimed: float
    claim_status: str
    claim_date: date
    resolved_date: Optional[date] = None
    claim_aging: Optional[int] = None

    class Config:
        orm_mode = True

class InventoryHealthResponse(BaseModel):
    product_id: str
    total_quantity_shipped: int
    total_claims: int
    claim_rate: float

    class Config:
        orm_mode = True

class NewShipmentRequest(BaseModel):
    origin_warehouse: str
    destination_city: str
    ship_date: date
    product_id: str
    quantity: int
    freight_cost: float

class NewShipmentResponse(BaseModel):
    shipment_id: str
    message: str

class UploadDeliveryLogsResponse(BaseModel):
    filename: str
    status: str
    message: str
    rows_processed: int
```

## Develop fastapi endpoints

### Subtask:
Develop fastapi endpoints


**Reasoning**:
Instantiate the FastAPI application, create a SQLAlchemy session dependency, and implement the specified API endpoints.



```python
from fastapi import FastAPI, Depends, HTTPException, File, UploadFile
from sqlalchemy.orm import Session
from sqlalchemy import func
from typing import List, Dict, Any
import pandas as pd
from io import BytesIO
from datetime import datetime

from __main__ import engine, Shipment, DeliveryLog, Claim
from __main__ import ClaimsSummaryResponse, InventoryHealthResponse, NewShipmentRequest, NewShipmentResponse, UploadDeliveryLogsResponse

app = FastAPI()

def get_db():
    db = Session(bind=engine)
    try:
        yield db
    finally:
        db.close()

@app.get("/claims-summary", response_model=List[ClaimsSummaryResponse])
def get_claims_summary(db: Session = Depends(get_db)):
    claims = db.query(Claim).all()
    claims_summary = []
    for claim in claims:
        claim_aging = (claim.resolved_date - claim.claim_date).days if claim.resolved_date else (datetime.now().date() - claim.claim_date).days
        claims_summary.append(ClaimsSummaryResponse(
            claim_id=claim.claim_id,
            delivery_id=claim.delivery_id,
            reason=claim.reason,
            amount_claimed=claim.amount_claimed,
            claim_status=claim.claim_status,
            claim_date=claim.claim_date,
            resolved_date=claim.resolved_date,
            claim_aging=claim_aging
        ))

    # Calculate claim percentages per carrier (requires joining with delivery_logs)
    claim_percentages = db.query(
        DeliveryLog.carrier,
        func.count(Claim.claim_id).label('total_claims'),
        func.count(DeliveryLog.delivery_id).label('total_deliveries')
    ).join(Claim, DeliveryLog.delivery_id == Claim.delivery_id).group_by(DeliveryLog.carrier).all()

    # This part is not directly covered by the ClaimsSummaryResponse model, but useful for analysis
    carrier_claim_percentage = {
        carrier: (total_claims / total_deliveries * 100) if total_deliveries > 0 else 0
        for carrier, total_claims, total_deliveries in claim_percentages
    }
    print("Carrier Claim Percentages:")
    print(carrier_claim_percentage)

    return claims_summary

@app.get("/inventory-health", response_model=List[InventoryHealthResponse])
def get_inventory_health(db: Session = Depends(get_db)):
    # Retrieve data and calculate total quantity shipped and total claims per product
    inventory_data = db.query(
        Shipment.product_id,
        func.sum(Shipment.quantity).label('total_quantity_shipped'),
        func.count(Claim.claim_id).label('total_claims')
    ).outerjoin(DeliveryLog, Shipment.shipment_id == DeliveryLog.shipment_id).outerjoin(Claim, DeliveryLog.delivery_id == Claim.delivery_id).group_by(Shipment.product_id).all()

    inventory_health_summary = []
    for product_id, total_quantity_shipped, total_claims in inventory_data:
        claim_rate = (total_claims / total_quantity_shipped * 100) if total_quantity_shipped > 0 else 0
        inventory_health_summary.append(InventoryHealthResponse(
            product_id=product_id,
            total_quantity_shipped=total_quantity_shipped,
            total_claims=total_claims,
            claim_rate=claim_rate
        ))
    return inventory_health_summary


@app.post("/log-shipment", response_model=NewShipmentResponse)
def log_new_shipment(shipment: NewShipmentRequest, db: Session = Depends(get_db)):
    # Generate a simple shipment ID
    last_shipment = db.query(Shipment).order_by(Shipment.shipment_id.desc()).first()
    if last_shipment:
        last_id_num = int(last_shipment.shipment_id[1:])
        new_id_num = last_id_num + 1
        new_shipment_id = f"S{new_id_num:05d}"
    else:
        new_shipment_id = "S00001"

    db_shipment = Shipment(
        shipment_id=new_shipment_id,
        origin_warehouse=shipment.origin_warehouse,
        destination_city=shipment.destination_city,
        ship_date=shipment.ship_date,
        # Assuming estimated_delivery_date is not provided in the request, set to ship_date for now
        delivery_date=shipment.ship_date, # Placeholder, this should be estimated or calculated
        product_id=shipment.product_id,
        quantity=shipment.quantity,
        freight_cost=shipment.freight_cost
    )
    db.add(db_shipment)
    db.commit()
    db.refresh(db_shipment)
    return NewShipmentResponse(shipment_id=db_shipment.shipment_id, message="Shipment logged successfully")

@app.post("/upload-delivery-logs", response_model=UploadDeliveryLogsResponse)
async def upload_delivery_logs(file: UploadFile = File(...), db: Session = Depends(get_db)):
    if not file.filename.endswith('.csv'):
        raise HTTPException(status_code=400, detail="Invalid file format. Only CSV files are accepted.")

    try:
        contents = await file.read()
        df = pd.read_csv(BytesIO(contents))

        # Basic validation: check for required columns
        required_columns = ['delivery_id', 'shipment_id', 'carrier', 'status', 'delivery_duration_days', 'damage_flag', 'proof_of_delivery_status']
        if not all(col in df.columns for col in required_columns):
             raise HTTPException(status_code=400, detail=f"Missing required columns. Required columns are: {', '.join(required_columns)}")

        rows_processed = 0
        for index, row in df.iterrows():
            # Check if delivery_id already exists to avoid duplicates
            existing_log = db.query(DeliveryLog).filter(DeliveryLog.delivery_id == row['delivery_id']).first()
            if existing_log is None:
                db_delivery_log = DeliveryLog(
                    delivery_id=row['delivery_id'],
                    shipment_id=row['shipment_id'],
                    carrier=row['carrier'],
                    status=row['status'],
                    delivery_duration_days=row['delivery_duration_days'],
                    damage_flag=bool(row['damage_flag']), # Ensure boolean type
                    proof_of_delivery_status=row['proof_of_delivery_status']
                )
                db.add(db_delivery_log)
                rows_processed += 1
        db.commit()

        return UploadDeliveryLogsResponse(filename=file.filename, status="success", message="Delivery logs uploaded successfully", rows_processed=rows_processed)

    except Exception as e:
        db.rollback()
        raise HTTPException(status_code=500, detail=f"An error occurred during file processing: {e}")
```
